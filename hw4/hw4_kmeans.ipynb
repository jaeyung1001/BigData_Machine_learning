# KMeans clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

## set clusters = 8

```python
est = KMeans(n_clusters=8)
print(est)
```

```python
est.fit(x)
```

```python
labels = est.labels_
print(labels)
```

```python
est.predict(x)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(x[:, 3], x[:, 0], x[:, 2], c=labels.astype(np.float), edgecolor='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')
ax.set_title("clusters 8")
ax.dist = 12
```

```python
# 2D
plt.figure()
plt.clf()
plt.scatter(x[:,0], x[:,1], c=labels, cmap=plt.cm.Set1,
           edgecolor='k')
# plt.label
plt.title('2D clustering result')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

## set clusters = 3

```python
est = KMeans(n_clusters=3)
est.fit(x)
labels = est.labels_
print(est.predict(x))
```

```python
Y_pred = np.choose(labels, [1, 0, 2]).astype(np.int64)
Y_pred
```

```python
# 2D
plt.figure(1, figsize=(8,6))
plt.clf()
plt.scatter(x[:,0], x[:,1], c=Y_pred, cmap=plt.cm.Set1,
           edgecolor='k')
# plt.label
plt.title('2D clustering result')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.figure(2, figsize=(8,6))
plt.clf()
plt.scatter(x[:,0], x[:,1], c=y, cmap=plt.cm.Set1,
           edgecolor='k')
plt.title('2D Ground truth result')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

## 정확도 측정

```python
count = 0
for i in range(len(Y_pred)):
    if Y_pred[i]== y[i]:
        count += 1

print("Accuracy is : {}%".format(round(count/len(Y_pred)*100,2)))
```

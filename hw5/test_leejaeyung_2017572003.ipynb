## 2번문제

```python
#2번문제
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
df = pd.read_csv('data\dataSet.csv', names=range(1,1001))
df.describe()
```

```python
df
```

## 테이블변환

```python
from pandas import Series,DataFrame
y = df.ix[7]
data = {'Y':y,'X1':df.ix[0],'X2':df.ix[1],'X3':df.ix[2],'X4':df.ix[3],'X5':df.ix[4],'X6':df.ix[5],'X7':df.ix[6]}
frame = DataFrame(data)
frame
```

```python
from sklearn import linear_model
from sklearn.cross_validation import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_predict
X = frame[frame.columns[0:6]]
Y = frame['Y']
reg = linear_model.LinearRegression()
predicted = cross_val_predict(reg, X,Y, cv=10)

print("MSE : {}".format(mean_squared_error(predicted,Y)))
fig, ax = plt.subplots()
ax.scatter(Y, predicted, edgecolors=(0,0,0))
ax.plot([Y.min(),Y.max()], [Y.min(),Y.max()],'k--',lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

```python
predict_result = DataFrame({"Predicted value":predicted,"True value":Y})
predict_result
```

# Correlation coefficient 계산

```python
corrcoef = []
# frame['Y']
for col in frame.columns:
    value = abs(numpy.corrcoef(frame['Y'],frame[col])[0][1])
    corrcoef.append([col,value])
    
corrcoef.sort(key=lambda x: x[1],reverse=True)

for value in corrcoef:
    print("for col", value[0],":\t",round(value[1],5))
```

```python
import pandas as pd
from pandas import Series, DataFrame
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('exercise.csv')
frame = DataFrame(data)
```

```python
frame = frame.set_value(frame[frame.D=='I'].index,'D','1')
frame = frame.set_value(frame[frame.D=='F'].index,'D','0')
frame
```

```python
feature = frame.L.values.reshape(-1,1)
label = frame.D
lm = LogisticRegression()
lm.fit(feature,label)
print("{:.2f}".format(lm.score(feature, label)*100))
```

```python
from sklearn.model_selection import KFold
for i in range(4,10):
    print('{}-fold cross validation..'.format(i))
    kf = KFold(n_splits=i, shuffle=True, random_state=0)
    for train_index, test_index in kf.split(feature):
        lm.fit(feature[train_index],label[train_index])
    #     print(lm.predict(feature[test_index]))
        print("{:.2f}".format(lm.score(feature[test_index], label[test_index])*100))
```
